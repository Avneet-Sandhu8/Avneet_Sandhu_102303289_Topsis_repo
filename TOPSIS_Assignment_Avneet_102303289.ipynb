```python
df = pd.read_csv("data.csv")
df
```

Step : Convert categorical to numeric(Already numeric in our dataset)

Why we do this:

1.First column is just names

2.TOPSIS math works only on numbers

```python
alternatives = df.iloc[:, 0]     # Mobile names
X = df.iloc[:, 1:]               # Criteria values (P1,P2,P3,P4,P5)

X
```

Step 2.1: Vector Normalization

Calculate Root of Sum of Squares

```python
X = df.iloc[:, 1:]   # criteria values
root_sum_squares = np.sqrt((X ** 2).sum())

root_sum_squares
```

Step 2.2: Vector Normalization

Normalized Decision Matrix

```python
normalized_decision_matrix = X / root_sum_squares
normalized_decision_matrix
```

Step 3.1: Weight Assignment

```python
weights = np.array([0.20, 0.20, 0.20, 0.20, 0.20])
weights
```

Step 3.2: Weight Assignment

Weighted Normalized Decision Matrix

```python
weighted_normalized_decision_matrix = normalized_decision_matrix * weights
weighted_normalized_decision_matrix
```

Step 4: Find Ideal Best and Ideal Worst

```python
impacts = ["-", "+", "+", "+","-"]

ideal_best = []
ideal_worst = []

for i in range(len(impacts)):
    if impacts[i] == "+":
        ideal_best.append(weighted_normalized_decision_matrix.iloc[:, i].max())
        ideal_worst.append(weighted_normalized_decision_matrix.iloc[:, i].min())
    else:
        ideal_best.append(weighted_normalized_decision_matrix.iloc[:, i].min())
        ideal_worst.append(weighted_normalized_decision_matrix.iloc[:, i].max())

ideal_best = np.array(ideal_best)
ideal_worst = np.array(ideal_worst)

ideal_best, ideal_worst
```

Step 5: Calculate Euclidean Distance

```python
S_plus = np.sqrt(((weighted_normalized_decision_matrix - ideal_best) ** 2).sum(axis=1))
S_minus = np.sqrt(((weighted_normalized_decision_matrix - ideal_worst) ** 2).sum(axis=1))

S_plus, S_minus
```

Step 6: Calculate Performance Score

```python
performance_score = S_minus / (S_plus + S_minus)
performance_score
```

Step 7: TOPSIS Score and Rank

```python
final_result = pd.DataFrame({
    "Mobile": df.iloc[:, 0],
    "Performance Score (Pi)": performance_score,
    "Rank": performance_score.rank(ascending=False).astype(int)
})

final_result.sort_values("Rank")
```
